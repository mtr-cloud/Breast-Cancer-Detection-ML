# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Defining names to columns of unlabelled dataset.

```python
names=['id','Clump Thickness','Cell Size','Cell Shape','Marginal Adhesion','Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin','Normal Nucleoli','Mitoses','Class']
```

```python
df=pd.read_csv("Data.csv",names=names)
```

# Splitting data into X(feature vector) and y(output vector)

```python
X=df.iloc[:,:-1].values
```

```python
X
```

```python
y=df.iloc[:,-1].values
```

```python
y
```

# Splitting data into training and test sets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

# Feature Scaling

```python
df=df.replace('?',-99999)
```

```python
df.head()
```

```python
df=df.replace('?',-9999)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train.reshape(-1,1)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
print(X)
```

```python
df['Clump Thickness'] = df['Clump Thickness'].astype('category')
df['Cell Size'] = df['Cell Size'].astype('category')
df['Cell Shape'] = df['Cell Shape'].astype('category')
df['Marginal Adhesion']=df['Marginal Adhesion'].astype('category')
df['Single Epithelial Cell Size'] = df['Single Epithelial Cell Size'].astype('category')
df['Bare Nuclei'] = df['Bare Nuclei'].astype('category')
df['Bland Chromatin'] = df['Bland Chromatin'].astype('category')
df['Normal Nucleoli'] = df['Normal Nucleoli'].astype('category')
df['Mitoses'] = df['Mitoses'].astype('category')
```

```python
df.dtypes
```

```python
X_train[:, 1:] = sc.fit_transform(X_train[:, 1:])
X_test[:, 1:] = sc.transform(X_test[:, 1:])
```

```python
print(X_train)
```

```python
print(X_test)
```

# Now applying KNN to the endoded data.

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

# Predicting a result from dataframe data as input

```python
df
```

```python
X_train[1,:]
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Checking for the accuracy for KNN model.

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

```python
print(accuracy_score(y_test,y_pred))
```

# Thus by using KNN we get an accuracy score of 95.43%.

# Now using SVM 

```python
from sklearn.svm import SVC
```

```python
classifier=SVC(kernel='linear',random_state=0)
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

# Using SVM the accuracy is improved a bit-96%.

# Now we will try using radial basis function kernel SVM and check for any improvement in model with same dataset and same features.

```python
from sklearn.svm import SVC
classifier_kernel=SVC(kernel='rbf',random_state=0)
classifier_kernel.fit(X_train,y_train)
```

```python
y_pred=classifier_kernel.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

# The efficiency is not improved.So using linear kernel gives better results than using rbf kernel.

# Naive Bayes.

# Using Decision tree classification

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

# Using Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

# Efficiency is improved with the same parameters(96.57%)

Thus we can conclude:
Using different classifiers-KNN,SVM(with linear,rbf kernels),Naive Bayes(Guassian kernel),Decision tree and random forest classification we have got the best accuracy with random forest classifier model-96.57%.So using this classifier can be a positive insight in medical science for prediction of breast cancer.
